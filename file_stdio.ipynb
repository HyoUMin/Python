```python
# 파일 생성하고 닫기
# 쓰기 모드('w')로 파일을 열 때 이미 존재하는 파일을 열면 그 파일의 내용이 모두 사라짐.
f = open('새파일.txt', 'w')
f.close()
```

```python
# 파일을 쓰기 모드로 열어 내용 쓰기
f = open('새파일.txt', 'w')
for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# readline 함수 사용하기 -> 파일에서 한 번에 한 줄씩 읽어오기
f = open('새파일.txt','r')
line = f.readline()
print(line)
f.close()
```

```python
# readlines 함수 사용하기 -> 파일에서 한 번에 모두 읽어오기. 리스트로 반환
f = open('새파일.txt','r')
lines = f.readlines()
for line in lines:
    line = line.strip()
    print(line)
f.close()
```

```python
# read 함수 사용하기 -> 파일 전체를 문자열로 반환
f = open('새파일.txt', 'r')
data = f.read()
print(data)
f.close()
```

```python
# for문 사용하기
f = open('새파일.txt', 'r')
for line in f:
    line = line.strip()
    print(line)
f.close()
```

```python
# 파일에 새로운 내용 추가하기
# 쓰기 모드로 기존에 있는 파일을 열면 덮어씌어져서 기존 내용은 사라짐.
# 추가 모드 ('a')를 사용해야 함.
f = open('새파일.txt', 'a')
for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# with문과 함께 사용하기
with open('test.txt', 'w') as F:
    F.write("Life is too short, you need python.") 
```
