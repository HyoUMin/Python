```python
a = {1: 'a'}
a[2] = 'b'
print(a)
# 딕셔너리 쌍 추가하기
```

```python
a = {1: 'a', 2: 'b', 'name': 'pey', 3: [1, 2, 3]}
del a[1]
print(a)
# 딕셔너리 a에서 key가 1인 쌍을 삭제하기.
```

```python
grade = {'pey': 10, 'julliet': 99}
print(grade['pey'])
print(grade['julliet'])
# key값으로 value 얻기
```

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(a.keys())
# key 리스트 만들기 -> 형변환 필요함.
print(list(a.keys()))
print()

print(a.values())
# value 리스트 만들기 -> 형변환 필요함.
print(list(a.values()))
print()

print(a.items())
# 쌍으로 얻기
print(list(a.items()))

a.clear()
# 모두 지우기
```

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
a.get('name')
# key를 통해 value  얻기
```

```python
a = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print('name' in a)
# 해당 key가 딕셔너리 안에 있는지 조사하기(True/False)
```

```python
a = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
print(a.pop('phone'))
# 해당 key로 value 반환함. 반환된 key-value는 삭제됨. 
```
