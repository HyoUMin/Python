```python
"{0:<10}".format("hi")
# 총 10칸, '<'이므로 왼쪽으로 정렬함.
```

```python
"{0:>5}".format("Hi")
# 총 5칸, '>'이므로 오르쪽으로 정렬함.
```

```python
"{0:^15}".format("Hi")
# 총 15칸, '^'이므로 가운데로 정렬함.
```

```python
"{0:!^7}".format("Hi")
# 총 7칸, '^'이므로 가운데로 정렬함. 
# '!'로 공백을 채움. '!'는 다른 문자로 변경가능함.
```

```python
"{0:10.4f}".format(10)
# 총 자릿수 10자리, 소숫점 4자리까지 표현하기 
```

```python
s = "12345"
s.isdigit()
# 문자열이 숫자로만 구성되어 있는지 확인하기(True/False)
```
