```python
a = [1, 2, 3]
del a[1] # 리스트 a의 인덱스 1번에 해당하는 2 삭제함.
# del로 리스트 요소 삭제하기
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:] # 리스트 a의 인덱스 2번부터 끝까지 삭제함.
print(a)
```

```python
a = [1, 2, 3]
a.append(4) # 리스트 a의 마지막 인덱스[2] 뒤에 4를 추가함.
print(a)
# 리스트 마지막에 요소 추가하기.
```

```python
a = [1, 4, 3, 2]
a.sort()
print(a)
# 리스트 정렬하기
```

```python
a = ['a', 'c', 'b']
a.reverse()
print(a)
# 현재 리스트를 거꾸로 뒤집기. 정렬하지 않음.
```

```python
a = [1, 2, 3]
print(a.index(3))
# 3에 해당하는 인덱스 번호를 반환함.
```

```python
a = [1, 2, 3]
a.insert(0, 4)
print(a)
# 리스트에 요소 삽입하기. 0번 인덱스 자리에 4를 삽입
```

```python
a = [1, 2, 3, 1, 2, 3]
a.remove(3)
print(a)
# 리스트에서 첫 번째로 나오는 3을 제거
```

```python
a = [1, 2, 3]
a.pop()
print(a)
# 리스트의 마지막 요소를 반환함. 반환한 요소는 리스트에서 삭제됨.
```

```python
a = [1, 2, 3]
a.pop(1)
print(a)
# 1번 인덱스에 해당하는 요소를 반환함. 반환한 요소는 리스트에서 삭제됨.
```

```python
a = [1, 2, 3, 1]
print(a.count(1))
# 리스트에 포함된 1의 개수 세기
```

```python
a = [1, 2, 3]
a.extend([4, 5])
print(a)
# 리스트 확장하기. 리스트 맨 뒤에 추가함.
```
